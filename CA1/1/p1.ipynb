1-Generate Binary Image

```python
import cv2
import numpy as np

roseimg = cv2.imread("rose.jpg")
roseimg = cv2.imread('rose.jpg', cv2.IMREAD_GRAYSCALE)

binaryThreshold = 90
binimage = roseimg

for i in range (0, roseimg.shape[0]):
        for index in range (0, roseimg.shape[1]):
            if roseimg[i][index] <= binaryThreshold:
                binimage[i][index] = 0
            else:
                binimage[i][index] = 255

cv2.imwrite("binrose.jpg", binimage)
```

Generate Grayscale

```python
import cv2

roseimg = cv2.imread("rose.jpg")
grayrose = cv2.cvtColor(roseimg, cv2.COLOR_BGR2GRAY)
cv2.imwrite("grayrose.jpg", grayrose)
```

Increase All Gray Pixels By 20

```python
for i in range(grayrose.shape[0]):
    for j in range(grayrose.shape[1]):
        grayrose[i, j] = grayrose[i, j] + 20

cv2.imwrite("plus20.jpg", grayrose)
```

Increase Contrast of RGB & GrayScale

```python
import cv2
import numpy as np

roseimg = cv2.imread('rose.jpg', cv2.IMREAD_GRAYSCALE)

median = np.median(roseimg)
print(median)
min = np.min(roseimg)
print(min)
max = np.max(roseimg)
print(max)

alpha = 2
beta = 15

highcontrastgray = roseimg.copy()

for i in range(roseimg.shape[0]):
    for j in range(roseimg.shape[1]):

        if roseimg[i, j] < median:
            highcontrastgray[i, j] = -roseimg[i, j] * alpha + beta                

        elif roseimg[i, j] > median:
            highcontrastgray[i, j] = roseimg[i, j] * alpha + beta

        else:
            highcontrastgray[i, j] = roseimg[i, j]

cv2.imwrite("highcontrastgray.jpg", highcontrastgray)
```

Decrease Contrast of GrayScale Image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 

roseimg = cv2.imread('rose.jpg', cv2.IMREAD_GRAYSCALE)

median = np.median(roseimg)
print(median)
min = np.min(roseimg)
print(min)
max = np.max(roseimg)
print(max)

alpha = 1/2
beta = 5

lowcontrastgray = roseimg.copy()

for i in range(roseimg.shape[0]):
    for j in range(roseimg.shape[1]):

        if roseimg[i, j] < median:
            lowcontrastgray[i, j] = -roseimg[i, j] * alpha + beta                

        elif roseimg[i, j] > median:
            lowcontrastgray[i, j] = roseimg[i, j] * alpha + beta

        else:
            lowcontrastgray[i, j] = roseimg[i, j]

cv2.imwrite("lowcontrastgray.jpg", lowcontrastgray)
```

Increase Contrast of RGB Image

```python
import cv2 
import matplotlib.pyplot as plt 
import numpy as np 
  
image = cv2.imread('rose.jpg') 
  
plt.subplot(1, 2, 1) 
plt.title("Original") 
plt.imshow(image) 

brightness = 10 
contrast = 2.3  

image2 = cv2.addWeighted(image, contrast, np.zeros(image.shape, image.dtype), 0, brightness) 

cv2.imwrite('highrgb.jpg', image2) 
plt.subplot(1, 2, 2) 
plt.title("Brightness & contrast") 
plt.imshow(image2) 
plt.show()
```

Decrease Contrast of RGB Image

```python
import cv2 
import matplotlib.pyplot as plt 
import numpy as np 
  
image = cv2.imread('rose.jpg') 
  
plt.subplot(1, 2, 1) 
plt.title("Original") 
plt.imshow(image) 

brightness = -5
contrast = 0.4

image2 = cv2.addWeighted(image, contrast, np.zeros(image.shape, image.dtype), 0, brightness) 

cv2.imwrite('lowrgb.jpg', image2) 
plt.subplot(1, 2, 2) 
plt.title("Brightness & contrast") 
plt.imshow(image2) 
plt.show()
```

Salt and Pepper Noise

```python
import random 
import cv2 
  
def add_noise(img): 
    row , col = img.shape 
    number_of_pixels = 150 * 150
    for i in range(number_of_pixels): 
        y_coord=random.randint(0, row - 1) 
        x_coord=random.randint(0, col - 1) 
        img[y_coord][x_coord] = 255

    number_of_pixels = 130 * 130
    for i in range(number_of_pixels): 
        y_coord=random.randint(0, row - 1) 
        x_coord=random.randint(0, col - 1) 
        img[y_coord][x_coord] = 0
          
    return img 

img = cv2.imread('grayrose.jpg', cv2.IMREAD_GRAYSCALE) 
cv2.imwrite('saltandpepper.jpg', add_noise(img)) 
```
